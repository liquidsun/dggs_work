```python
import pandas as pd
import numpy as np
```

```python
import os

def read_stats(filename, results_dir, projection):
    fh = os.path.join(results_dir, filename)
    df = pd.read_csv(fh)
    df['projection'] = projection
    df['meta'] = filename
    return df

all_confs = [
# isea

("cell_stats_DGGRID_ISEA3H", "results", "isea"),
("cell_stats_DGGRID_ISEA4D", "results", "isea"),
("cell_stats_DGGRID_ISEA4H", "results", "isea"),
("cell_stats_h3", "results", "isea"),
("cell_stats_rhpix", "results", "isea"),
("cell_stats_s2", "results", "isea"),

# goode

("cell_stats_DGGRID_ISEA3H", "results1", "goode"),
( "cell_stats_DGGRID_ISEA4D", "results1", "goode"),
( "cell_stats_DGGRID_ISEA4H", "results1", "goode"),
("cell_stats_h3", "results1", "goode"),
("cell_stats_rhpix", "results1", "goode"),
( "cell_stats_s2", "results1", "goode")
    
]

all_df = []

for d in all_confs:
    df = read_stats(d[0], d[1], d[2])
    all_df.append(df)

df1 = pd.concat(all_df, axis=0)

for att in ['min_area','max_area','std','mean']:
    df1[f"{att}_km2"] = df1[att] / 1000000

df1.sample(5)

```
